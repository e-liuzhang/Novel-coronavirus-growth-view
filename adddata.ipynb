```python
import csv

# 读取csv至字典
with open("data1.csv", "r",encoding='utf-8') as csvFile:
    reader = csv.reader(csvFile)
    result = {}
    value=[]
    name=[]
    date=[]
    for item in reader:
        # 忽略第一行
        if reader.line_num == 1:
            continue
        name.append(item[0])
        date.append(item[2])
        # 判断是否为空
        if item[1]:
            # 除去负数
            if int(item[1])>=0:
                # 累加
                try:
                    result[item[0]] =int(result[item[0]])+int(item[1])
                except Exception as e:
                    result[item[0]] = item[1]
                value.append(result[item[0]])
            else:
                value.append(result[item[0]])
        else:
            value.append(result[item[0]])

fileHeader = ["name", "value","date"]
with open("data2.csv", "w",encoding='utf-8',newline='') as fileChange:
    writer = csv.writer(fileChange)
    # 填头
    writer.writerow(fileHeader)
    for n in range(0,len(name)):
        writer.writerow([name[n],value[n],date[n]])  
    
    
```
